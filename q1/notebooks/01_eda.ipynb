```python
# eda.py
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("G:/My Drive/github/msc-ml-cw/q1/data/bank-additional-full.csv", sep=";")  # adjust path if needed
```

```python
# -------------------
# 1. Basic Overview
# -------------------
print("Shape of dataset:", df.shape)
print("\nColumns:\n", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum().sort_values(ascending=False))
```

```python
# Show first few rows
print("\nFirst 5 rows:\n", df.head())
```

```python
# Check for null values
print("\nNull values in each column:\n", df.isnull().sum())
```

```python
# Number of each target class
print("\nTarget variable distribution:\n", df['y'].value_counts())
```

```python
# -------------------
# 2. Target Variable Distribution
# -------------------
plt.figure(figsize=(6,4))
sns.countplot(x="y", data=df, palette="viridis")
plt.title("Target Variable Distribution")
plt.xlabel("Subscription (y)")
plt.ylabel("Count")
plt.show()
```

```python
# -------------------
# 3. Numerical Features Distribution
# -------------------
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[num_cols].hist(bins=30, figsize=(15,10), color="steelblue", edgecolor="black")
plt.suptitle("Numerical Feature Distributions")
plt.show()
```

```python
# -------------------
# 4. Categorical Features Distribution
# -------------------
cat_cols = df.select_dtypes(include=["object"]).columns.drop("y")

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index, palette="Set2")
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# -------------------
# 5. Correlation Heatmap (numerical features only)
# -------------------
plt.figure(figsize=(10,8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```

```python
# -------------------
# 6. Relationship between numerical features & target
# -------------------
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x="y", y=col, data=df, palette="Set1")
    plt.title(f"{col} vs Target")
    plt.show()
```

```python
# -------------------
# 7. Relationship between categorical features & target
# -------------------
for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, hue="y", data=df, palette="Set1")
    plt.title(f"{col} vs Target")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Class imbalance
imbalance_ratio = df['y'].value_counts(normalize=True)
print(f"\nClass imbalance:\n{imbalance_ratio}")
```

```python
# Outliers using IQR
for col in num_cols:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    outliers = ((df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)).sum()
    print(f"{col}: {outliers} outliers")
```

```python
# pdays special value analysis
print(f"\nNever contacted (pdays=999): {(df['pdays']==999).sum()}")
```

```python
# Unknown values in categoricals
for col in cat_cols:
    if 'unknown' in df[col].values:
        print(f"{col} unknowns: {(df[col]=='unknown').sum()}")
```

```python
# Success rate by categorical features
for col in cat_cols:
    print(f"\n{col} success rates:")
    print(df.groupby(col)['y'].apply(lambda x: (x=='yes').mean()).sort_values(ascending=False))
```

```python
# 1. Pairplot for key numerical features
key_nums = ['age', 'campaign', 'previous', 'emp.var.rate', 'euribor3m']
sns.pairplot(df[key_nums + ['y']], hue='y', diag_kind='kde')
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```

```python
# 2. Temporal patterns
plt.figure(figsize=(10,5))
month_order = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
df['y_binary'] = (df['y'] == 'yes').astype(int)
df.groupby('month')['y_binary'].mean().reindex(month_order).plot(kind='line', marker='o')
plt.title("Success Rate by Month")
plt.ylabel("Conversion Rate")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
df.groupby('day_of_week')['y_binary'].mean().plot(kind='bar')
plt.title("Success Rate by Day of Week")
plt.ylabel("Conversion Rate")
plt.show()
```

```python
# 3. Campaign intensity vs success
plt.figure(figsize=(10,5))
df.groupby('campaign')['y_binary'].mean().plot(kind='line', marker='o')
plt.title("Success Rate vs Number of Contacts (campaign)")
plt.xlabel("Number of Contacts")
plt.ylabel("Conversion Rate")
plt.grid(True)
plt.show()
```

```python
# 4. Previous outcome impact
plt.figure(figsize=(8,4))
df.groupby('poutcome')['y_binary'].mean().plot(kind='bar')
plt.title("Success Rate by Previous Outcome")
plt.ylabel("Conversion Rate")
plt.show()
```

```python
# 5. Economic indicators variance check
econ_features = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
print("\nEconomic indicators - unique values:")
for col in econ_features:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
# 6. Duration analysis (even though it will be dropped)
plt.figure(figsize=(8,4))
sns.boxplot(x='y', y='duration', data=df)
plt.title("Call Duration vs Target (Note: Data Leakage)")
plt.show()
```
